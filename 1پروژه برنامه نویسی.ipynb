```python
import math
def hesab_panel(wat,dehi,shodan):
    wat_inverter = wat*125/100
    i = 100
    v = 12
    p = i * v #توان هر باتری
    meghdar_batri = math.ceil(dehi/(wat_inverter/p)) #تعداد باتری های مورد نیاز با استفاده است مقدار زمان شارژدهی داده شده
    jaryan_kol_batri = i * meghdar_batri #جریان کل باتری ها
    jaryan_khrooji_panel = math.ceil(jaryan_kol_batri / shodan) #جریان خروجی از پنل
    jaryan_khrooji_panel_satandard = math.ceil(jaryan_kol_batri /5) #جریان استاندارد خروجی از پنل تا به باتری آسیبی وارد نشود
    
    if jaryan_khrooji_panel <= jaryan_khrooji_panel_satandard :
        meghdar_panel = math.ceil(jaryan_khrooji_panel / 5) #تعداد پنل مورد نیاز بر اساس جریان
    else:
        print('مقدار زمان شارژ شدن کم تر از حد معمول است و ممکن است به باتری آسیب برساند')
        meghdar_sharzh_shodan_standard=jaryan_khrooji_panel_satandard / jaryan_kol_batri
        print('زمان میانگین شارژ شدن در حالت استاندار:',meghdar_sharzh_shodan_standard)
        meghdar_panel = math.ceil(jaryan_khrooji_panel_satandard / 5) #تعداد پنل مورد نیاز بر اساس جریان استاندارد
        
    print('هر باتری : [100Ah , 12V , {}W ] :1000تومان \n هر پنل: [5A , 12V , 60W] : 1500تومان \n اینورتر:به ازای هر 100 وات 300تومان '.format(p))
    print('تعداد پنل:{} \t  تعداد باتری:{} \t وات اینورتر:{}'.format(meghdar_panel,meghdar_batri,wat_inverter))
    gheimat_panel = meghdar_panel * 1500
    gheimat_batri = meghdar_batri * 1000
    gheimat_inverter = wat_inverter/100*300
    gheimat_kol = gheimat_panel + gheimat_batri + gheimat_inverter
    print('قیمت کل:',gheimat_kol)
    
    return gheimat_kol
hesab_panel(1000,10,5)
```
